## EDA

```python
plt.figure(figsize = (3, 5))
sns.countplot(data = data1, x = 'gender')
plt.title('Gender')
```

성비는 남자가 더 많다

```python
sns.set_palette(sns.color_palette(['#ED553B', '#3CAEA3']))
plt.figure(figsize = (10, 4))
sns.countplot(data = data1, x = 'age_group', hue = 'gender',
              order = data1['age_group'].value_counts().index)
plt.title('Age & Gender')
```

```python
data1['age_group'].value_counts()#.sort_index().index
```

```python
data1.loc[data1['gender'] == 'Male']['age_group'].value_counts().sort_index().reset_index()['age_group'].tolist()
```

```python
data1['age_group'].value_counts().sort_index().reset_index()['index'].tolist()
```

```python
labels = data1['age_group'].value_counts().sort_index().reset_index()['index'].tolist()
mens = data1.loc[data1['gender'] == 'Male']['age_group'].value_counts().sort_index().reset_index()['age_group'].tolist()
womens = data1.loc[data1['gender'] == 'Female']['age_group'].value_counts().sort_index().reset_index()['age_group'].tolist()
# men_std = [2, 3, 4, 1, 2]
# women_std = [3, 5, 2, 3, 3]
width = 0.6       # the width of the bars: can also be len(x) sequence

fig, ax = plt.subplots()

ax.bar(labels, mens, width, label='Men')
ax.bar(labels, womens, width, bottom=mens,
       label='Women')

ax.set_ylabel('count')
ax.set_title('by age group and gender')
ax.legend()

plt.show()
```

10대와 20대의 비율이 월등히 높은것을 알 수 있다.
